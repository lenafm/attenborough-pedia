## Take the combined link file and generate noun > adj, bird > adj, noun > noun and adj > adj files

```python
# imports
import os
import json
```

```python
# read the json files
nouns = json.load(open("output/nouns.json"))
adjs = json.load(open("output/adjs.json"))
birds = json.load(open("output/birds.json"))

links = json.load(open("output/links.json"))
```

```python
"pelican" in nouns
```

```python
# bird adj link generation
bird_adjs = []

for link in links:
  if link[0] in birds and link[1] in adjs:
    bird_adjs.append([link[0], link[1]])
  if link[1] in birds and link[0] in adjs:
    bird_adjs.append([link[1], link[0]])
    
with open('output/bird_adjs.json', 'w') as f:
  f.write(json.dumps(bird_adjs))
```

```python
# noun adj link generation
noun_adjs = []

for link in links:
  if link[0] in nouns and link[1] in adjs:
    noun_adjs.append([link[0], link[1]])
  if link[1] in nouns and link[0] in adjs:
    noun_adjs.append([link[1], link[0]])
    
with open('output/noun_adjs.json', 'w') as f:
  f.write(json.dumps(noun_adjs))

noun_adjs
```

```python
# noun noun link generation
noun_nouns = []

for link in links:
  if link[0] in nouns and link[1] in nouns and link[1] != link[0]:
    noun_nouns.append([link[0], link[1]])
    
with open('output/noun_nouns.json', 'w') as f:
  f.write(json.dumps(noun_nouns))

noun_nouns
```
